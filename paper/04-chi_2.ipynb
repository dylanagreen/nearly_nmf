```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
chi_2_noisy_nnls = np.load("data/chi_2_noisy_nnls.npy")
chi_2_noiseless_nnls = np.load("data/chi_2_noiseless_nnls.npy")
```

```python
plt.rcParams.update({'font.size': 18})
fig, ax = plt.subplots(figsize=(10, 5), layout="constrained")

diff_nnls = chi_2_noisy_nnls - chi_2_noiseless_nnls
print(np.mean(diff_nnls), np.median(diff_nnls))
# bins = np.linspace(np.min(diff_nnls), np.max(diff_nnls), 80)
bins = np.linspace(-1.5, 1.5, 81)

plt.hist(diff_nnls, bins=bins, histtype="step", label="$\chi^2_{noisy} - \chi^2_{noisefree}$")
plt.title("Nearly-NMF with NNLS coefficients")
plt.xlabel("$\chi^2$ Difference")
plt.grid(alpha=0.5)

plt.axvline(x=np.mean(diff_nnls), c="C1", alpha=0.7, label="Mean Difference")

# plt.yscale("log")
plt.xlim(-1.5, 1.5)
plt.ylabel("Num Quasars")
plt.legend()
plt.savefig("plots/chi_2_diff.pdf", bbox_inches="tight")
```

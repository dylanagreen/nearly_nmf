```python
import matplotlib.pyplot as plt
import numpy as np
# plt.rcParams['text.usetex'] = True
```

```python
nearly_chi_2 = np.load("nearly_chi_2.npy")
shift_chi_2 = np.load("shift_chi_2.npy")
```

```python
plt.rcParams.update({'font.size': 16})
fig, ax = plt.subplots(figsize=(15, 8))
plt.plot(nearly_chi_2.T, c="C0", alpha=0.1)
plt.plot(shift_chi_2.T, c="C1", alpha=0.1, ls="dotted")

plt.ylabel("$\chi^2$", dict(fontsize=30))
plt.xlabel(r"Iteration Number", dict(fontsize=20))

ylim = ax.get_ylim()

# These two won't show up they're just for the legend.
plt.plot([-1], [-1], c="C1", ls="dotted", label="Shift-NMF")
plt.plot([-1], [-1], c="C0", label="Nearly-NMF")
ax.set(ylim=ylim, xlim=(0, 30))#, yscale="log")
plt.grid()

ax.legend()

plt.savefig("plots/convergence.pdf", bbox_inches="tight", dpi=256)
```
